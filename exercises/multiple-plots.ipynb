# Multiple plots

With the `.subplot()` method we can draw multiple plots in the same figure.

This method takes 3 parameters that describe the layout of the figure. The layout is organized in rows and columns, which are represented by the first two parameters. The third parameter represents the index that the current plot occupies.

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
# plot 1
fig, (ax1, ax2) = plt.subplots(2, 1)
l1 = np.array([3, 5, 1, 9])

plt.subplot(1, 2, 1)
plt.plot(l1, c = "magenta", ls = ":", lw = 2)

# plot 2
l2 = np.array([8, 2, 6, 4])

plt.subplot(1, 2, 2)
plt.plot(l2, c = "blue", ls = "--", lw = 2)

plt.show()
```

**Observation** In the above example, `plt.subplot(1, 2, 1)` indicates that the layout has one row, 2 columns and that the plot to be drawn next occupies the first plot. Whereas `plt.subplot(1, 2, 2)`, again indicates that the layout consists of one row and two columns, but in this case the subsequent plot will be drawn on the second plot.

If, on the other hand, we want a layout of the figure that consists of 2 rows and a column, we would do the following:

```python
# plot 1
plt.subplot(2, 1, 1)
plt.plot(l1, c = "magenta", ls = ":", lw = 2)

# plot 2
plt.subplot(2, 1, 2)
plt.plot(l2, c = "blue", ls = "--", lw = 2)

plt.show()
```

We can draw as many subplots as we want inside a figure. It will always be necessary to indicate the number of rows, the number of columns and the index that each plot occupies.

**Attention!** In this case, the indices of the plots go from 1 to $ n $, where $ n $ = rows $ \ times $ columns of the layout. The first index corresponds to the cell in the first row and first column; the second, with first row second column; the third corresponds to the box in the first row, third column; and so on until the first row is completed. Then it continues with the second row and so on until all the rows have been completed.

#### Example 1

We are going to make a layout that consists of 2 rows and 4 columns. In total we will have 8 subplots, so that the indices will go from 1 to 8:

```python
plt.subplots(figsize = (12, 8))

# plot 1 (1st row 1st column)
plt.subplot(2, 4, 1)
plt.plot(l1, c = "blue", ls = "-", lw = 2)

# plot 2 (1st row 2nd column)
plt.subplot(2, 4, 2)
plt.plot(l2, c = "magenta", ls = "--", lw = 2)

# plot 3 (1st row 3rd column)
plt.subplot(2, 4, 3)
plt.plot(l1, c = "black", ls = ":", lw = 2)

# plot 4 (1st row 4rth column)
plt.subplot(2, 4, 4)
plt.plot(l2, c = "cyan", ls = "-.", lw = 2)

# plot 5 (2nd row 1st column)
plt.subplot(2, 4, 5)
plt.plot(l2, c = "black", ls = "--", lw = 2)

# plot 6 (2nd row 2nd column)
plt.subplot(2, 4, 6)
plt.plot(l1, c = "cyan", ls = "-.", lw = 2)

# plot 7 (2nd row 3rd column)
plt.subplot(2, 4, 7)
plt.plot(l2, c = "blue", ls = "-", lw = 2)

# plot 8 (2nd row 4rth col)
plt.subplot(2, 4, 8)
plt.plot(l1, c = "magenta", ls = ":", lw = 2)

plt.show()
```
